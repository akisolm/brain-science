```python
import pandas as pd

# å‡è®¾åŽŸå§‹ DataFrame å·²ç»åŠ è½½ä¸º dfï¼Œä¸”åŒ…å«åˆ— ['Yp','Zp_0','NSAp','NCIPp']

# 1. å®šä¹‰å–å€¼èŒƒå›´
years    = sorted(df['Yp'].unique())
z_groups = [0, 1]            # 0 = z<0, 1 = z>0
sa_vals  = list(range(1, 7))   # SA æ•°ï¼š1â€“6
cip_vals = list(range(1, 10))  # CIP æ•°ï¼š1â€“9

# 2. ç»Ÿè®¡æ¯ (Yp, Zp_0) çš„æ€»è®ºæ–‡æ•°
totals = (
    df
    .groupby(['Yp', 'Zp_0'])
    .size()
    .rename('total')
    .reset_index()
)

# 3. SA ç»´åº¦ï¼šæ¯ (Yp, Zp_0, NSAp) çš„è®ºæ–‡æ•°
counts_sa = (
    df
    .groupby(['Yp', 'Zp_0', 'NSAp'])
    .size()
    .rename('count')
    .reset_index()
)

# 4. æž„é€ å®Œæ•´ç´¢å¼•å¹¶è¡¥é½ SA
idx_sa = pd.MultiIndex.from_product(
    [years, z_groups, sa_vals],
    names=['Yp', 'Zp_0', 'NSAp']
)
sa_df = (
    pd.DataFrame(index=idx_sa)
      .reset_index()
      .merge(counts_sa, how='left', on=['Yp','Zp_0','NSAp'])
      .merge(totals,    how='left', on=['Yp','Zp_0'])
      .fillna({'count':0, 'total':0})
)
sa_df['frac'] = sa_df['count'] / sa_df['total']

# 5. CIP ç»´åº¦ï¼šæ¯ (Yp, Zp_0, NCIPp) çš„è®ºæ–‡æ•°
counts_cip = (
    df
    .groupby(['Yp', 'Zp_0', 'NCIPp'])
    .size()
    .rename('count')
    .reset_index()
)

# 6. æž„é€ å®Œæ•´ç´¢å¼•å¹¶è¡¥é½ CIP
idx_cip = pd.MultiIndex.from_product(
    [years, z_groups, cip_vals],
    names=['Yp', 'Zp_0', 'NCIPp']
)
cip_df = (
    pd.DataFrame(index=idx_cip)
      .reset_index()
      .merge(counts_cip, how='left', on=['Yp','Zp_0','NCIPp'])
      .merge(totals,     how='left', on=['Yp','Zp_0'])
      .fillna({'count':0, 'total':0})
)
cip_df['frac'] = cip_df['count'] / cip_df['total']

# 7. ç»“æžœ
# sa_df åŒ…å«åˆ— ['Yp','Zp_0','NSAp','count','total','frac']ï¼Œå¯¹åº”ä¸Šå›¾è¦ç”»çš„æ•°æ®
# cip_df åŒ…å«åˆ— ['Yp','Zp_0','NCIPp','count','total','frac']ï¼Œå¯¹åº”ä¸­å›¾è¦ç”»çš„æ•°æ®

# ä½ å¯ä»¥ç›´æŽ¥æŸ¥çœ‹ï¼š
# print(sa_df.head())
# print(cip_df.head())
```

```python
import pandas as pd
import json

# å‡è®¾ df å·²ç»åœ¨çŽ¯å¢ƒä¸­åŠ è½½ï¼ŒåŒ…å« ['Yp','Zp_0','NSAp','NCIPp']

# 0. å¯¹å¤§äºŽ4çš„ç±»åˆ«è¿›è¡Œæˆªæ–­
df['NSAp']  = df['NSAp'].clip(upper=4)
df['NCIPp'] = df['NCIPp'].clip(upper=4)

# 1. å®šä¹‰å¹´ä»½å’Œç±»åˆ«èŒƒå›´
years    = sorted(df['Yp'].unique())
z_groups = [0, 1]            # 0 = ä½Žå¼•ç”¨, 1 = é«˜å¼•ç”¨
sa_vals  = [1, 2, 3, 4]      # åªä¿ç•™å‰4ä¸ªSAç±»åˆ«
cip_vals = [1, 2, 3, 4]      # åªä¿ç•™å‰4ä¸ªCIPç±»åˆ«

# 2. ç»Ÿè®¡æ¯å¹´æ¯å¼•ç”¨ç»„æ€»è®ºæ–‡æ•°
totals = df.groupby(['Yp','Zp_0']).size().rename('total').reset_index()

# 3. ç»Ÿè®¡â€œæ°å¥½ n ä¸ª SAâ€æ–‡ç« æ•°
counts_sa = (
    df.groupby(['Yp','Zp_0','NSAp'])
      .size()
      .rename('count')
      .reset_index()
)

# 4. è¡¥é½æ‰€æœ‰ (Yp, Zp_0, n) ç»„åˆï¼Œè®¡ç®— frac
idx_sa = pd.MultiIndex.from_product([years, z_groups, sa_vals], names=['Yp','Zp_0','NSAp'])
sa_df = (
    pd.DataFrame(index=idx_sa).reset_index()
      .merge(counts_sa, how='left', on=['Yp','Zp_0','NSAp'])
      .merge(totals,    how='left', on=['Yp','Zp_0'])
      .fillna({'count':0,'total':0})
)
sa_df['frac'] = sa_df['count'] / sa_df['total']

# 5. åŒç†å¤„ç† CIP
counts_cip = (
    df.groupby(['Yp','Zp_0','NCIPp'])
      .size()
      .rename('count')
      .reset_index()
)
idx_cip = pd.MultiIndex.from_product([years, z_groups, cip_vals], names=['Yp','Zp_0','NCIPp'])
cip_df = (
    pd.DataFrame(index=idx_cip).reset_index()
      .merge(counts_cip, how='left', on=['Yp','Zp_0','NCIPp'])
      .merge(totals,    how='left', on=['Yp','Zp_0'])
      .fillna({'count':0,'total':0})
)
cip_df['frac'] = cip_df['count'] / cip_df['total']

# 6. å¯¼å‡ºä¸º JSON ä¾› D3.js ä½¿ç”¨
def df_to_d3_json(df, key_col, prefix, vals):
    result = []
    for (year, z), grp in df.groupby(['Yp','Zp_0']):
        obj = {'year': int(year), 'z': int(z)}
        for n in vals:
            key = f"{prefix}_{n}"
            obj[key] = float(grp.loc[grp[key_col]==n, 'frac'])
        result.append(obj)
    return result

sa_json  = df_to_d3_json(sa_df,  'NSAp',  'sa',  sa_vals)
cip_json = df_to_d3_json(cip_df, 'NCIPp', 'cip', cip_vals)

# å°† JSON ä¿å­˜åˆ°æ–‡ä»¶
with open('sa_data.json',  'w') as f: json.dump(sa_json,  f, indent=2)
with open('cip_data.json', 'w') as f: json.dump(cip_json, f, indent=2)
```
